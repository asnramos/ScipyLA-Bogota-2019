```python
import os
import folium

print(folium.__version__)
```

```python
import numpy as np

data = (np.random.normal(size=(25, 1)) * np.array([[.25, .25, .25]]) + np.array([-24.7907199,-65.4163422, .25])).tolist()
```

```python
from folium.plugins import HeatMap

m = folium.Map([-24.7907199,-65.4163422], control_scale = True, tiles='stamentoner', zoom_start=7)

HeatMap(data).add_to(m)

m.save(os.path.join('results', 'Heatmap.html'))

m
```
